# Contoh 2
Gunakan Python untuk memplot fungsi sin bersama dengan aproksimasi deret Taylor orde pertama, ketiga, kelima, dan ketujuh. Perhatikan bahwa ini melibatkan suku nol hingga ketiga dalam rumus yang diberikan sebelumnya.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-poster")
```

```python
x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ["Orde Pertama", "Orde Kedua", "Orde Kelima", "Orde Ketujuh"]

plt.figure(figsize = (10,8))
for n, label in zip(range(4), labels):
    y=y+((-1)**n*(x)**(2*n+1))/np.math.factorial(2*n+1)
    plt.plot(x,y, label = label)
    
plt.plot(x, np.sin(x), "k", label = "Analytic")
plt.grid()
plt.title("Pendekatan Deret Taylor dari Berbagai Orde")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Contoh 1
Sebagai contoh pada bagian ini kita menentukan deret Taylor untuk fungsi sin(x).

```python
%matplotlib inline
from sympy import *
import numpy as np
```

Menentukan koefisien deret Taylor.
Menggunakan metode diff dan factorial, menghitungnya secara berulang.

```python
x = symbols('x')
```

```python
fx = sin(x)
```

```python
coef = []
for n in range(10):
    fnx = diff(fx,x,n)
    cn = fnx.subs(x,0)/factorial(n)
    coef.append(cn)
coef
```
