# Fungsi Menghitung Luas dan Volume Bola

```python
from numpy import pi

def rumus_bola(r):
    V = (4/3)*pi*r**3       # volume bola
    A = 4*pi*r**2           # luas bola
    return V,A
```

```python
rumus_bola(4)
```

# Module Persamaan Kuadrat dan Kubik (file persamaan harus extensi .py)

```python
import matplotlib.pyplot as plt

from numpy import arange
from persamaan import kuadrat, kubik       # memanggil fungsi kuadrat dan kubik module persamaan (ada di file persamaan.py)

x = arange(-20.0,20.0,0.1)
y1 = kuadrat(x)
y2 = kubik(x)

plt.subplot(3,1,1)                          # posisi grafik yaitu pada posisi urut 1 dengan baris 3 kolom 1
plt.plot(x,y1,'-')                         # sumbu x dan y serta bentuk kurva berupa garis
plt.title('GRAFIK PERSAMAAN KUADRAT')       # judul grafik
plt.xlabel('x')                             # label untuk sumbu x
plt.ylabel('y')                             # label sumbu y
plt.grid(True)                              # garis bantu grid


plt.subplot(3,1,3)                          # posisi grafik yaitu pada posisi urut 3 dengan baris 3 kolom 1
plt.plot(x,y2,'-')                         # sumbu x dan y serta bentuk kurva berupa garis
plt.title('GRAFIK PERSAMAAN KUBIK')         # judul grafik
plt.xlabel('x')                             # label untuk sumbu x
plt.ylabel('y')                             # label sumbu y
plt.grid(True)                              # garis bantu grid

plt.show()                                  # menampilkan plot
```
